```python
import wget
import os
```

```python
def create_folders(parent_folder, folder_names):
    # Join the parent folder with each folder name to create the full path
    folder_paths = [os.path.join(parent_folder, folder_name) for folder_name in folder_names]

    # Create the folders
    for folder_path in folder_paths:
        os.makedirs(folder_path, exist_ok=True)
        print(f"Folder created: {folder_path}")

# Example usage
parent_folder = '/portal1/dell/li-b/CGLS/'
folder_names = ['2014', '2015', '2016', '2017', '2018', '2019', '2020', '2021', '2022']

create_folders(parent_folder, folder_names)
```

```python
# Function to generate URLs for the specified date range
def generate_urls(start_year, end_year):
    base_url = "https://land.copernicus.vgt.vito.be/PDF/datapool/Vegetation/Properties/LAI_300m_V1"
#     template = "{}/{:02d}/{:02d}/LAI300_{}{:02d}{:02d}0000_GLOBE_PROBAV_V1.0.1/c_gls_LAI300_{}{:02d}{:02d}0000_GLOBE_PROBAV_V1.0.1.nc"
    template = "{}/{:02d}/{:02d}/LAI300-RT6_{}{:02d}{:02d}0000_GLOBE_OLCI_V1.1.2/c_gls_LAI300-RT6_{}{:02d}{:02d}0000_GLOBE_OLCI_V1.1.2.nc" #2021
    # template = "{}/{:02d}/{:02d}/LAI300-RT6_{}{:02d}{:02d}0000_GLOBE_OLCI_V1.1.2/c_gls_LAI300-RT6_{}{:02d}{:02d}0000_GLOBE_OLCI_V1.1.2.nc" #2022
    	
    	
    urls = []
    for year in range(start_year, end_year + 1):
        for month in range(5, 13):
            # Assuming last day of the month is used
            day = 28 if month == 2 else 31 if month in [1, 3, 5, 7, 8, 10, 12] else 30
            # day = 10
            # day = 20
            url = template.format(
                year, month, day,
                year, month, day,
                year, month, day
            )
            urls.append(f"{base_url}/{url}")

    return urls

# Generate URLs for the year range 2014 to 2020
start_year = 2021
end_year = 2021
urls = generate_urls(start_year, end_year)

# Write URLs to a text file
output_file = "/portal1/dell/li-b/CGLS/url2021111122222.txt"
with open(output_file, "w") as file:
    file.write("\n".join(urls))

print(f"URLs written to {output_file}")
```
